```python
import pandas as pd
from annoy import AnnoyIndex
import pickle

print("Iniciando pipeline de exportación para Producción...")

# 1. Cargar el dataset original limpio
ruta_csv = "C:/Users/basti/proyecto_cd/data/processed/spotify_dataset_processed.csv"
df = pd.read_csv(ruta_csv)

features = ['danceability', 'energy', 'loudness', 'speechiness', 
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

dimensiones = len(features)
annoy_model = AnnoyIndex(dimensiones, 'angular')

# 2. Alimentar y construir el modelo
for i in range(len(df)):
    vector_acustico = df.iloc[i][features].values
    annoy_model.add_item(i, vector_acustico)

annoy_model.build(10)

# 3. EXPORTAR EL MODELO FÍSICO
annoy_model.save('modelo_spotify.ann')
print("✅ Modelo Annoy exportado exitosamente como 'modelo_spotify.ann'")

# 4. EXPORTAR EL CATÁLOGO LIGERO (Para el módulo de explicabilidad)
# Solo guardamos el ID y las variables acústicas, nada de textos pesados
df_minimal = df[['track_id'] + features]
df_minimal.to_pickle('catalogo_ligero.pkl')
print("✅ Catálogo minimalista exportado como 'catalogo_ligero.pkl'")
```
